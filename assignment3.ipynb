```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Building_Inventory.csv",na_values = {'Year Acquired': 0, 'Year Constructed': 0, 'Square Footage': 0, 'Total Floors': 0})
```

```python
df.plot(x='Year Acquired', y='Year Constructed', kind='scatter', title='Relationship between the year acquired and the year constructed')
```

The above scatter plot show very clearly about the rough correlation between two variables and also gives the comprehensive distribution of two variable including outliers.

Shortcomings: 
When the data is too much, as there are so many dots just cover each other so we are not able to see the exact density.

```python
y = df.groupby("Congress Dist")["Square Footage"].sum()
```

```python
x = list(df.groupby("Congress Dist").groups.keys())
```

```python
plt.bar(x, y)
xi = [x[i] for i in range(len(x))]
plt.xticks(xi, x)
plt.xlabel("Congressional District")
plt.ylabel("Total Square Footage")
plt.title("Total square footage as a function vs Congressional district")
```

The above bar graph shows the total square footage for each district 
and we can directly compare them by the visualization.

Shortcomings: The change of square footage along with the district is not as clear as using line plot.

```python
df_tmp = df
df_tmp["square footage per floor"]=df["Square Footage"]/df['Total Floors']
```

```python
z = df_tmp.groupby("Congress Dist")["square footage per floor"].mean()
plt.bar(x, z)
xi = [x[i] for i in range(len(x))]
plt.xticks(xi, x)
plt.xlabel("Congressional District")
plt.ylabel("Average Square Footage per Floor")
plt.title("Average square footage per floor as a function VS Congressional district")
```

The above graph shows the Average Square Footage per Floor for 
each district and we can directly compare them by the visualization.
 
Shortcomings:
The change of square footage along with the district is not as clear as using line plot.
